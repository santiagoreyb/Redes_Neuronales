### Observar significancia de los atributos respecto a la calidad

```python
from sklearn import linear_model
import statsmodels.api as sm
from scipy import stats
```

```python
x = red_white[['fixed acidity','residual sugar','chlorides','total sulfur dioxide', 'density','pH','sulphates','alcohol','RedWine']]
y = red_white['quality']

# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)

X2 = sm.add_constant(x)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
y_pred = regr.predict(x)
plt.scatter(x['alcohol'], y_pred, color="royalblue")
```

## Eliminar datos atipicos

```python
boxplot = red_white.boxplot(column=['fixed acidity','residual sugar','alcohol'], return_type='both', figsize=(10,10))
plt.show()
```

```python
#Fixed acidity
Q1 = np.percentile(red_white['fixed acidity'], 25, method='midpoint')
Q3 = np.percentile(red_white['fixed acidity'], 75, method='midpoint')
IQR = Q3 - Q1

upper=Q3+1.5*IQR
lower=Q1-1.5*IQR
print(upper,'fixed acidity')
print(lower,'fixed acidity')
upper_array = np.where(red_white['fixed acidity']>=upper)[0]
lower_array = np.where(red_white['fixed acidity']<=lower)[0]
print(upper_array)
print(lower_array)
print("Observaciones antes: ", red_white['fixed acidity'].count())
red_white.drop(index=upper_array,axis=0,inplace=True, errors='ignore')
red_white.drop(index=lower_array ,inplace=True, errors='ignore')
print("Observaciones despues: ", red_white['fixed acidity'].count())

#Residual sugar

Q1 = np.percentile(red_white['residual sugar'], 25, method='midpoint')
Q3 = np.percentile(red_white['residual sugar'], 75, method='midpoint')
IQR = Q3 - Q1

upper=Q3+1.5*IQR
lower=Q1-1.5*IQR
print(upper,'residual sugar')
print(lower,'residual sugar')
upper_array1 = np.where(red_white['residual sugar']>=upper)[0]
lower_array1 = np.where(red_white['residual sugar']<=lower)[0]
print(upper_array1)
print(lower_array1)
print("Observaciones antes: ", red_white['residual sugar'].count())
red_white.drop(index=upper_array1,axis=0, inplace=True, errors='ignore')
red_white.drop(index=lower_array1, inplace=True, errors='ignore')
print("Observaciones despues: ", red_white['residual sugar'].count())


#alcohol

Q1 = np.percentile(red_white['alcohol'], 25, method='midpoint')
Q3 = np.percentile(red_white['alcohol'], 75, method='midpoint')
IQR = Q3 - Q1

upper=Q3+1.5*IQR
lower=Q1-1.5*IQR
print(upper,'alcohol')
print(lower,'alcohol')
upper_array2 = np.where(red_white['alcohol']>=upper)[0]
lower_array2 = np.where(red_white['alcohol']<=lower)[0]
print(upper_array2)
print(lower_array2)
print("Observaciones antes: ", red_white['alcohol'].count())
red_white.drop(index=upper_array2,axis=0, inplace=True, errors='ignore')
red_white.drop(index=lower_array2, inplace=True, errors='ignore')
print("Observaciones despues: ", red_white['alcohol'].count())


```
